Resize and rescale video frames

```python
import cv2 as cv 


def rescaleFrame(frame, scale=0.75):
    # Used for images , videos or live videos
    width = int(frame.shape[1] * scale)
    height = int(frame.shape[0] * scale)
    dimensions = (width , height)

    return cv.resize(frame , dimensions , interpolation=cv.INTER_AREA)


def changeRes(width , height):
    # Live video 
    capture.set(3 , width)
    capture.set(4 , height)


# reading videos 
capture = cv.VideoCapture(r"C:\Users\chatt\Desktop\Computer Vision Tutorials\OpenCV Tutorials\Resources\Videos\dog.mp4")

while True:
    isTrue , frame = capture.read() 

    frame_resized = rescaleFrame(frame)

    cv.imshow("Video", frame)
    cv.imshow("Resized_Video" , frame_resized)
    if cv.waitKey(20) & 0xFF == ord('d'):
        break

capture.release()
cv.destroyAllWindows()
```
